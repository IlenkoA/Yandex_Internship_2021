```python
import pandas as pd
df = pd.read_excel('sql_data.xlsx')
df2 = df[(df['Category'] == 'ПОТРЕБИТЕЛЬСКОЕ КРЕДИТОВАНИЕ') & (df['Month']>='2020-01-01') & (df['Month']<='2020-08-01') & (df['Cost_rub_wo_NDS']!=0)]
df1 = df[(df['Client'] == 'Омега') & (df['Category'] == 'ПОТРЕБИТЕЛЬСКОЕ КРЕДИТОВАНИЕ') & (df['Month']>='2020-01-01') & (df['Month']<='2020-08-01')]
df2
```

```python
df2['CPA_adj'] = df2['Сonversions'] / df2['Cost_rub_wo_NDS']
df2['CPM'] = df2['Cost_rub_wo_NDS'] / df2['Shows'] * 1000
df2['CPC'] = df2['Cost_rub_wo_NDS'] / df2['Clicks']
df2
```

```python
#df1
df2.Client.value_counts()
```

```python
df3 = df2.groupby('Client')['Cost_rub_wo_NDS'].sum().reset_index(name = 'Total')
df4 = df3.sort_values(by = ['Total'], ascending = False).round()
```

```python
import matplotlib.pylab as plt
import numpy as np
fig = plt.figure(figsize = (12, 8))
plt.bar(df4.Client, df4.Total)
plt.ylabel('Costs, 10^8 rub', fontsize = 15)
plt.tick_params(labelsize = 12)
plt.show()
```

### Теперь сделаем сравнение эффективности, используя суммирование вместо AVG

```python
df6 = df2.groupby('Client')[['Cost_rub_wo_NDS','Shows','Clicks', 'Сonversions']].sum().reset_index().sort_values(by = ['Cost_rub_wo_NDS'], ascending = False)
df6
```

```python
df6['CPA'] = df6['Cost_rub_wo_NDS'] / df6['Сonversions']
df6['CPM'] = df6['Cost_rub_wo_NDS'] / df6['Shows'] * 1000
df6['CPC'] = df6['Cost_rub_wo_NDS'] / df6['Clicks']
df6.sort_values(by = ['CPA'], ascending = True)
```

```python
df7 = df6[['Client','CPA','CPC']].sort_values(by = ['CPC'], ascending = True).round(decimals=1)
df7
```

```python
fig = plt.figure(figsize = (12, 8))
plt.bar(df7.Client, df7.CPA)
plt.ylabel('CPA', fontsize = 15)
plt.tick_params(labelsize = 13)
plt.show()
```

```python
fig = plt.figure(figsize = (12, 8))
plt.bar(df7.Client, df7.CPC)
plt.ylabel('CPC', fontsize = 15)
plt.tick_params(labelsize = 13)
plt.show()
```
